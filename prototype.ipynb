```python
# Improved sentiment feature extraction for better feature engineering
def enhanced_sentiment_features(text: str, max_tokens: int = 512):
    """
    Extracts a rich set of sentiment features from text using FinBERT and keyword heuristics.

    Returns a tuple:
        - sentiment_score: +1 (POS), -1 (NEG), 0 (NEU)
        - sentiment_confidence: confidence of majority sentiment
        - sentiment_polarity: (pos - neg) / total chunks
        - has_negative_keywords: 1 if negative keywords present, else 0
        - word_count: number of words in text
        - pos_frac, neg_frac, neu_frac: fraction of each sentiment in chunks
    """
    if pd.isna(text) or not isinstance(text, str) or text.strip() == "":
        return (
            0,      # sentiment_score
            0.0,    # sentiment_confidence
            0.0,    # sentiment_polarity
            0,      # has_negative_keywords
            0,      # word_count
            0.0,    # pos_frac
            0.0,    # neg_frac
            0.0     # neu_frac
        )

    # Clean text
    text_clean = re.sub(r"\s+", " ", text)
    text_clean = re.sub(r"http\S+|www\S+", "", text_clean)
    text_clean = text_clean.strip()

    # Negative keywords heuristic
    negative_keywords = ['struggle', 'difficult', 'challenge', 'stress', 'debt', 'late', 'miss', 'problem', 'overdue', 'hardship', 'default']
    has_negative = int(any(word in text_clean.lower() for word in negative_keywords))
    word_count = len(text_clean.split())

    # Tokenize and chunk for long text
    tokens = tokenizer(text_clean, return_tensors="pt", truncation=False)
    input_ids = tokens["input_ids"][0]
    num_chunks = (len(input_ids) + max_tokens - 1) // max_tokens

    sentiments = []
    confidences = []
    for i in range(num_chunks):
        chunk_ids = input_ids[i * max_tokens:(i + 1) * max_tokens]
        if len(chunk_ids) < 5:
            continue
        chunk_text = tokenizer.decode(chunk_ids, skip_special_tokens=True)
        try:
            result = classifier(chunk_text, truncation=True, max_length=max_tokens)
            # result is a list of dicts, so take the first
            result = result[0]
            label = result["label"].upper()
            score = result["score"]
            sentiments.append(label)
            confidences.append(score)
        except Exception as e:
            logging.warning(f"Chunk error: {e}")
            continue

    if not sentiments:
        return (
            0,      # sentiment_score
            0.0,    # sentiment_confidence
            0.0,    # sentiment_polarity
            has_negative,
            word_count,
            0.0,    # pos_frac
            0.0,    # neg_frac
            0.0     # neu_frac
        )

    # Majority sentiment and its confidence
    from collections import Counter
    sentiment_counter = Counter(sentiments)
    majority = sentiment_counter.most_common(1)[0][0]
    # Get indices of majority sentiment
    majority_indices = [i for i, s in enumerate(sentiments) if s == majority]
    avg_conf = sum(confidences[i] for i in majority_indices) / len(majority_indices) if majority_indices else 0.0

    # Sentiment score mapping
    sentiment_score = 1 if majority == "POSITIVE" else -1 if majority == "NEGATIVE" else 0

    # Sentiment polarity and fractions
    pos_count = sentiments.count("POSITIVE")
    neg_count = sentiments.count("NEGATIVE")
    neu_count = sentiments.count("NEUTRAL")
    total = len(sentiments)
    polarity = (pos_count - neg_count) / total if total > 0 else 0.0
    pos_frac = pos_count / total if total > 0 else 0.0
    neg_frac = neg_count / total if total > 0 else 0.0
    neu_frac = neu_count / total if total > 0 else 0.0

    return (
        sentiment_score,
        round(avg_conf, 4),
        round(polarity, 4),
        has_negative,
        word_count,
        round(pos_frac, 4),
        round(neg_frac, 4),
        round(neu_frac, 4)
    )
```

```python
import logging
from typing import Tuple
import re


def clean_text(text: str) -> str:
    text = re.sub(r"\s+", " ", text)  # remove excessive whitespace
    text = re.sub(r"http\S+|www\S+", "", text)  # remove URLs
    return text.strip()

# Make sure logging is set up
logging.basicConfig(level=logging.WARNING)

def get_sentiment_features(text: str, max_length: int = 512) -> Tuple[int, float]:
    """
    Analyze the sentiment of text using a FinBERT classifier.

    Parameters:
        text (str): Input text string.
        max_length (int): Max token length for model input (default=512).

    Returns:
        Tuple[int, float]: 
            - sentiment_value: 1 (positive), -1 (negative), 0 (neutral or invalid)
            - confidence: Model's confidence score (0.0 if failed)
    """
    if not isinstance(text, str) or not text.strip():
        return 0, 0.0

    try:
        cleaned_text = clean_text(text)
        result = classifier(cleaned_text, truncation=True, max_length=max_length)[0]

        label = result.get("label", "").upper()
        confidence = result.get("score", 0.0)

        # Optional: treat low-confidence as neutral
        if confidence < 0.5:
            return 0, confidence

        sentiment_value = {
            "POSITIVE": 1,
            "NEGATIVE": -1,
            "NEUTRAL": 0
        }.get(label, 0)

        return sentiment_value, confidence

    except Exception as e:
        logging.warning(f"Sentiment analysis failed: {e}")
        return 0, 0.0
```

```python
import xgboost as xgb
import numpy as np

# To avoid kernel crashes, subsample the data for initial model development
# (You can increase n_samples as memory allows, or remove this for full data)
n_samples = 50000  # Try 50k rows, adjust as needed for your RAM
if len(x_train) > n_samples:
    # Stratified sample to preserve class balance
    from sklearn.model_selection import StratifiedShuffleSplit
    sss = StratifiedShuffleSplit(n_splits=1, train_size=n_samples, random_state=42)
    idx, _ = next(sss.split(x_train, y_train))
    x_train_sub = x_train.iloc[idx].copy()
    y_train_sub = y_train.iloc[idx].copy()
else:
    x_train_sub = x_train.copy()
    y_train_sub = y_train.copy()

if len(x_test) > n_samples // 2:
    from sklearn.model_selection import StratifiedShuffleSplit
    sss = StratifiedShuffleSplit(n_splits=1, train_size=n_samples // 2, random_state=42)
    idx, _ = next(sss.split(x_test, y_test))
    x_test_sub = x_test.iloc[idx].copy()
    y_test_sub = y_test.iloc[idx].copy()
else:
    x_test_sub = x_test.copy()
    y_test_sub = y_test.copy()

# Calculate scale_pos_weight for class imbalance
n_pos = np.sum(y_train_sub == 1)
n_neg = np.sum(y_train_sub == 0)
scale_pos_weight = n_neg / n_pos if n_pos > 0 else 1

# Drop the columns with invalid dtypes and ignore missing columns
cols_to_drop = [
    'emp_title', 'url', 'zip_code', 'disbursement_method', 'title',
    'last_credit_pull_d', 'earliest_cr_line', 'last_pymnt_d', 'out_prncp', 'last_fico_range_high', 'id'
]
x_train_valid = x_train_sub.drop(columns=cols_to_drop, errors='ignore').copy()
x_test_valid = x_test_sub.drop(columns=cols_to_drop, errors='ignore').copy()

# Convert object columns that cause issues to the 'category' type
categorical_cols = [
    'grade', 'issue_d', 'pymnt_plan', 'initial_list_status',
    'hardship_flag', 'debt_settlement_flag', 'FICO_rating'
]
for col in categorical_cols:
    if col in x_train_valid.columns and x_train_valid[col].dtype == 'object':
        x_train_valid[col] = x_train_valid[col].astype('category')
    if col in x_test_valid.columns and x_test_valid[col].dtype == 'object':
        x_test_valid[col] = x_test_valid[col].astype('category')

# Remove any columns with dtype 'object' that are not handled
object_cols_train = x_train_valid.select_dtypes(include=['object']).columns
object_cols_test = x_test_valid.select_dtypes(include=['object']).columns
x_train_valid = x_train_valid.drop(columns=object_cols_train, errors='ignore')
x_test_valid = x_test_valid.drop(columns=object_cols_test, errors='ignore')

# Convert any pandas Sparse[bool] columns to dense bool (XGBoost does not support sparse bool)
for df_ in [x_train_valid, x_test_valid]:
    for col in df_.columns:
        if pd.api.types.is_sparse(df_[col]) and pd.api.types.is_bool_dtype(df_[col].dtype):
            df_[col] = df_[col].astype(bool)

# Reduce model complexity to avoid OOM: fewer trees, shallower trees, lower learning rate
xgb_1st = xgb.XGBClassifier(
    objective='binary:logistic',
    use_label_encoder=False,
    random_state=42,
    early_stopping_rounds=10,
    verbose=1,
    eval_metric='aucpr',
    enable_categorical=True,
    scale_pos_weight=scale_pos_weight,
    n_estimators=100,      # Lower number of trees
    max_depth=4,           # Shallower trees
    learning_rate=0.1,     # Reasonable learning rate
    tree_method='hist',    # More memory-efficient tree method
    subsample=0.8,         # Use a fraction of data per tree
    colsample_bytree=0.8   # Use a fraction of features per tree
)

xgb_1st.fit(
    x_train_valid, y_train_sub,
    eval_set=[(x_train_valid, y_train_sub), (x_test_valid, y_test_sub)]
)
```

```python
from sklearn.metrics import f1_score, precision_score


def evaluate_model(model, X_test, y_test, label="Model"):
    if hasattr(model, "feature_names_in_"):
        X_test = X_test.reindex(columns=model.feature_names_in_, fill_value=0)
    
    if len(X_test) != len(y_test):
        print(f"[ERROR] Shape mismatch: X_test has {len(X_test)} rows, y_test has {len(y_test)} rows.")
        return
    
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]

    print(f"\n===== {label} =====")
    print("Accuracy      :", accuracy_score(y_test, y_pred))
    print("Precision     :", precision_score(y_test, y_pred))
    print("Recall        :", recall_score(y_test, y_pred))
    print("F1 Score      :", f1_score(y_test, y_pred))
    print("ROC AUC Score :", roc_auc_score(y_test, y_proba))
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    print("="*50)
```

```python

from sklearn.metrics import RocCurveDisplay, PrecisionRecallDisplay
import matplotlib.pyplot as plt

def plot_comparisons(y_true, proba1, proba2, label1="With Sentiment", label2="Without Sentiment"):
    plt.figure(figsize=(12, 5))

    # PR Curve
    plt.subplot(1, 2, 1)
    PrecisionRecallDisplay.from_predictions(y_true, proba1, name=label1)
    PrecisionRecallDisplay.from_predictions(y_true, proba2, name=label2)
    plt.title("Precision-Recall Curve")

    # ROC Curve
    plt.subplot(1, 2, 2)
    RocCurveDisplay.from_predictions(y_true, proba1, name=label1)
    RocCurveDisplay.from_predictions(y_true, proba2, name=label2)
    plt.title("ROC Curve")

    plt.tight_layout()
    plt.show()
```
